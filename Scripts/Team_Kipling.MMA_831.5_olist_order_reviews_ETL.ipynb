```python
#this method will remove it. we will ignore the NAs for now, we will deal with them later
def deEmojify(inputString):
    if(pd.notna(inputString)):
        return inputString.encode('ascii', 'ignore').decode('ascii')
    
```

```python
#loop throught every row, send in the review comment and title into deEmojify method
for index,row in data.iterrows():
    #print(data.loc[index, 'review_comment_message'] )
    data.loc[index, 'review_comment_title'] = deEmojify(data.loc[index, 'review_comment_title'])
    data.loc[index, 'review_comment_message'] = deEmojify(data.loc[index, 'review_comment_message'])
```
