```python
##################### Creating Units Sold Metric
```

```python
#aggregating units sold by product
olist_products_sold = olist_order_items.groupby('product_id').count()[['order_id']]
#renaming order_id column
olist_products_sold.rename(columns = {'order_id':'units_sold'}, inplace = True)
#display top 5 rows
olist_products_sold.head()
```

```python
#joining olist_products_sold to olist_products
olist_products = pd.merge(olist_products, olist_products_sold, how = 'left', on = 'product_id')
olist_products.head()
```

```python
#checking for NULL values in units_sold column
olist_products.count() # NO NULL values
```

```python
##################### Creating Total Revenue per Product Metric
```

```python
olist_order_items.head()
```

```python
#aggregating price, by product, to get total revenue per product
olist_revenue = olist_order_items.groupby('product_id').sum()[['price']]
#renaming order_id column
olist_revenue.rename(columns = {'price':'revenue'}, inplace = True)
#display top 5 rows
olist_revenue.head()
```

```python
#joining olist_revenue to olist_products
olist_products = pd.merge(olist_products, olist_revenue, how = 'left', on = 'product_id')
olist_products.head()
```

```python
##################### Creating Average Price of Product Metric
```

```python
olist_products['avg_price'] = olist_products['revenue']/olist_products['units_sold']
olist_products.head()
```

```python
##################### Creating Average Rating Metric
```

```python
olist_order_reviews.head()
```

```python
#Check data type
print(olist_order_reviews.dtypes)
```

```python
#convert to datetime format
olist_order_reviews['review_creation_date'] = pd.to_datetime(olist_order_reviews['review_creation_date'], format = '%d-%m-%Y %H:%M')
olist_order_reviews['review_answer_timestamp'] = pd.to_datetime(olist_order_reviews['review_answer_timestamp'], format = '%d-%m-%Y %H:%M')
```

```python
#ordering olist_order_reviews by review creation date
olist_order_reviews = olist_order_reviews.sort_values(by ='review_creation_date' , ascending=True)
```

```python
#selecting latest review for each order
olist_order_reviews_latest = olist_order_reviews.loc[olist_order_reviews.groupby('order_id').review_creation_date.idxmax()]
olist_order_reviews_latest.head()
```

```python
#merging review score with olist_order_items table
olist_order_items = pd.merge(olist_order_items, olist_order_reviews_latest[['order_id','review_score']], how = 'left', on = 'order_id')
olist_order_items.head()
```

```python
#aggregate review score at product_id level
olist_product_review_score = olist_order_items.groupby('product_id').agg({'review_score':'sum','order_id':'count'})
#calculate average review score for each product
olist_product_review_score['avg_review_score'] = olist_product_review_score['review_score']/olist_product_review_score['order_id']
olist_product_review_score.head()
```

```python
#merging avg_review_score into olist_products table
olist_products = pd.merge(olist_products, olist_product_review_score[['avg_review_score']], how = 'left', on = 'product_id')
olist_products.head()
```

```python
##################### Creating Average Delivery Time Metric
```

```python
olist_orders.head()
```

```python
#Check data type
print(olist_orders.dtypes)
```

```python
#convert all data, time info into datetime format
olist_orders['order_purchase_timestamp'] = pd.to_datetime(olist_orders['order_purchase_timestamp'], format = '%Y-%m-%d %H:%M:%S')
olist_orders['order_approved_at'] = pd.to_datetime(olist_orders['order_approved_at'], format = '%Y-%m-%d %H:%M:%S')
olist_orders['order_delivered_carrier_date'] = pd.to_datetime(olist_orders['order_delivered_carrier_date'], format = '%Y-%m-%d %H:%M:%S')
olist_orders['order_delivered_customer_date'] = pd.to_datetime(olist_orders['order_delivered_customer_date'], format = '%Y-%m-%d %H:%M:%S')
olist_orders['order_estimated_delivery_date'] = pd.to_datetime(olist_orders['order_estimated_delivery_date'], format = '%Y-%m-%d %H:%M:%S')
```

```python
#adding delivery time column to olist_orders
olist_orders['delivery_time'] = (olist_orders['order_delivered_customer_date'] - olist_orders['order_purchase_timestamp'])
olist_orders['delivery_time_days'] = olist_orders['delivery_time'].dt.days + 1 # "+1" is to round up the number of days i.e. if an order took 25 days and 2hrs, delivery_time_days should show 26 days
olist_orders.head()
```

```python
#merge delivery_time_days with order_items table
olist_order_items = pd.merge(olist_order_items, olist_orders[['order_id','delivery_time_days']], how = 'left', on = 'order_id')
olist_order_items.head()
```

```python
#group delivery_time_days by product_id in olist_order_items table
olist_avg_delivery_time = olist_order_items.groupby('product_id').agg({'delivery_time_days':'sum', 'order_id':'count'})
olist_avg_delivery_time.head()
```

```python
#calculate avg_delivery_time_days for each product
olist_avg_delivery_time['avg_delivery_time_days'] = olist_avg_delivery_time['delivery_time_days']/olist_avg_delivery_time['order_id']
olist_avg_delivery_time.head()
```

```python
#merge avg_delivery_time_days to olist_products table
olist_products = pd.merge(olist_products, olist_avg_delivery_time[['avg_delivery_time_days']], how = 'left', on = 'product_id')
olist_products.head()
```

```python
##################### Creating No. of Sellers per Product Metric
```

```python
#count distinct seller_id's for each product_id
olist_sellers_per_product = olist_order_items.groupby('product_id').nunique()[['seller_id']]
#rename seller_id column to sellers
olist_sellers_per_product.rename(columns = {'seller_id':'sellers'}, inplace = True)
olist_sellers_per_product.head()
```

```python
#merge olist_sellers_per_product with olist_products table
olist_products = pd.merge(olist_products, olist_sellers_per_product, how = 'left', on = 'product_id')
olist_products.head()
```

```python
##################### Creating Units Sold per Seller per Product Metric
```

```python
olist_products['units_sold_per_seller'] = olist_products['units_sold']/olist_products['sellers']
olist_products.head()
```
